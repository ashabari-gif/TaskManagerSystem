```python
import pandas as pd
import bcrypt


df=pd.DataFrame(columns=["Username","Password"])
df.to_csv("taskManagerFile.csv",index=False)

df_task=pd.DataFrame(columns=["TaskDescription","Status","task_id"])
df_task.to_csv("taskFile.csv",index=False)

def checkUniqueName(name):
    try:
        # Read the CSV file
        df2 = pd.read_csv("taskManagerFile.csv")       
        # Check if the name is present in the 'Username' column
        is_present = name in df2['Username'].values
        
        # Return the result
        return is_present
    except FileNotFoundError:
        print("File not found. Please make sure 'taskManagerFile.csv' exists.")
        return False
    except KeyError:
        print("The 'Username' column does not exist in 'taskManagerFile.csv'.")
        return False
   
def hashPassword(password): 
    hashed_pwd=bcrypt.hashpw(password.encode('utf-8'),bcrypt.gensalt())
    print("nnnnnnn")
    print(hashed_pwd)
    return hashed_pwd

def login_func():
    print("Do you want to login? Y/N")
    login=input()
    if login.lower()=="y":
     while True:
        print("enter login username")
        login_name=input("").strip()
        is_present_uname=login_name in df['Username'].values
        if is_present_uname:
           print("login name is valid.")
           print("enter login password")
           login_pass=input("").strip()

           stored_password = df.loc[df['Username'] == login_name, 'Password'].values[0]
           isMatch=bcrypt.checkpw(login_pass.encode(), stored_password.encode())   
            
           if isMatch:
              print("Password match.")
              return True
           else:
             print("Incorrect password.")
             return False
        else:
           print("Invalid login name.Please enter valid login name!")
           break

def add_Task(): 

       while True:
        print("Add task description")
        task_desp=input("")
        task_status="Pending"       

        try:
            # Attempt to load the existing task file
            df_task = pd.read_csv("taskFile.csv", index_col=False)
            # Drop any unnamed columns that may have been added during CSV loading
            df_task = df_task.loc[:, ~df_task.columns.str.contains('^Unnamed')]
        except FileNotFoundError: 
            print("File not found. A new file will be created.")
            # Create an empty DataFrame with the necessary columns if file doesn't exist
            df_task = pd.DataFrame(columns=["TaskDescription", "Status", "task_id"])

        # Set the new task_id as the next number after the current max
        if not df_task.empty:
            next_id = df_task['task_id'].max() + 1
        else:
            next_id = 1

        # Create a DataFrame for the new task
        df_new_task = pd.DataFrame([[task_desp, task_status, next_id]], 
                                   columns=["TaskDescription", "Status", "task_id"])

        # Concatenate the new task to the existing DataFrame
        df_task = pd.concat([df_task, df_new_task], ignore_index=True)

        # Save the updated DataFrame back to CSV
        df_task.to_csv("taskFile.csv", index=False)
        print("Task added successfully:")
        print("******View task********")   
        #view task
        print(df_task)
        break        

def view_task():
    print(df_task)

def update_status():
    print("enter task id")
    data_task= pd.read_csv("taskFile.csv",index_col=False)
    taskId=int(input()) 
    data_task.loc[data_task['task_id'] == taskId, 'Status'] = 'Completed'
    data_task.to_csv("taskFile.csv", index=False)
    print("******Updated status*****")
    print(data_task)
    
def delete_completed_row():
    print("enter task id")
    data_task= pd.read_csv("taskFile.csv",index_col=False)
    taskId=int(input())  
    data_task.drop(data_task[data_task["task_id"]==taskId].index,inplace=True)
    #data_task=data_task[data_task['Status']!='Completed']
    print("****Task file after deleted row****")
    print(data_task)
           
```

```python

print("Do you want to registered? Y/N")
x=input()
if x.lower()=="y":
 while True:
   print("Enter Unique username.")
   u_name=input("") 
   if checkUniqueName(u_name):
     print("This name is already taken, please use unique Name!")
   else:
     print("Name is available.")
     print("Enter password")  
     p_word=input("") 
     hash_pwd=hashPassword(p_word) 
     print("Name registered successfully!")  
     df1=pd.DataFrame([[u_name,hash_pwd.decode('utf-8')]],columns=["Username","Password"]) 
     df = pd.concat([df,df1])
     df.to_csv("taskManagerFile.csv")  
     print(df)
     break  
```

```python
if login_func():
    print("Login is successfully!")
    print("Access to task manager")

    while True:
     print("Do you want to add task? y/n")
     add_task=input("")
     if add_task.lower()=="y":    
      add_Task()
     else:
      break
    while True:
      print("Do you want to update status? Y/N")
      status=input("")
      if status.lower()=="y":  
        update_status() 
      else:
        break  
    while True:
      print("Do you want to delete row? Y/N")
      status=input("")
      if status.lower()=="y":  
         delete_completed_row()  
      else:
        break              
else:
    print("login is unsuccessfully!")

```
